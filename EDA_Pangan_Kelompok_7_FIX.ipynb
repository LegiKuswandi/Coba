Hapus satuan dan jenis_produksi, mengganti nama kolom jumlah_produksi dan menghapus data dibawah tahun 2017 pada tabel sapi

```python
# #hapus satuan dan jenis_produksi
del dfsapi['satuan']
del dfsapi['jenis_produksi']

#rename
dfsapi = dfsapi.rename(columns={'jumlah_produksi': 'daging_sapi'})

#hapus data dibawah tahun 2017
dfsapi = dfsapi.loc[dfsapi['tahun'] >= 2017]

dfsapi.head(10)
```

Mengganti satuan pada tabel sapi, sebelumnya tabel sapi memiliki satuan kilogram sementara tabel yang lain memiliki satuan ton jadi disini kita memutuskan untuk menyamakan satuan nya menjadi ton.

```python
def kg_to_ton(daging_sapi):
    return daging_sapi / 1000

dfsapi['daging_sapi'] = dfsapi['daging_sapi'].apply(kg_to_ton)
dfsapi['daging_sapi'] = dfsapi['daging_sapi'].round(0).astype(int)

dfsapi.head(10)
```

Mengubah dfagama agar memiliki atribut kolom nama - nama dari agama, yang sebelumnya jadi isi dari kolom agama.

```python
# Pivot tabel untuk mengubah agama menjadi kolom baru
dfagama_baru = dfagama.pivot_table(index=['kode_provinsi', 'nama_provinsi', 'kode_kabupaten_kota', 'nama_kabupaten_kota', 'tahun'],
                               columns='agama', values='jumlah_penduduk', aggfunc='first')

# Reset indeks untuk menjadikan beberapa kolom sebagai indeks
dfagama_baru.reset_index(inplace=True)
dfagama_baru = dfagama_baru.loc[dfagama_baru['tahun']>=2017]
dfagama_baru = dfagama_baru.rename(columns={'agama': 'id'})
dfagama_baru.head(1000)
```

Menghapus kolom satuan dalam dfpengeluaran

```python
# #hapus satuan
del dfpengeluaran['satuan']

#hapus data dibawah tahun 2017
dfpengeluaran = dfpengeluaran.loc[dfpengeluaran['tahun'] >= 2017]

dfpengeluaran.head(10)
```

Karena tadi ada satuan ribu rupiah, tetapi tadi sudah kita hapus, jadi kita mengkalikan jumlah pengeluaran dengan 1000 agar menjadi jutaan 

```python
#mengubah satuan ribu rupiah jadi rupiah
def sejutain(jumlah_pengeluaran_per_kapita):
    return jumlah_pengeluaran_per_kapita * 1000

dfpengeluaran['jumlah_pengeluaran_per_kapita'] = dfpengeluaran['jumlah_pengeluaran_per_kapita'].apply(sejutain)

dfpengeluaran.head(1000)
```

```python
#mengubah tipe data jumlah_pengeluaran_per_kapita
dfpengeluaran['jumlah_pengeluaran_per_kapita'] = dfpengeluaran['jumlah_pengeluaran_per_kapita'].round(0).astype(int)
```

Menyatukan semua datadrame 

```python
dfm = dfayam.merge(dfbabi[['kode_kabupaten_kota', 'tahun', 'daging_babi']], on=['kode_kabupaten_kota', 'tahun'], how='left')
dfm = dfm.merge(dfdomba[['kode_kabupaten_kota', 'tahun', 'daging_domba']], on=['kode_kabupaten_kota', 'tahun'], how='left')
dfm = dfm.merge(dfkelinci[['kode_kabupaten_kota', 'tahun', 'daging_kelinci']], on=['kode_kabupaten_kota', 'tahun'], how='left')
dfm = dfm.merge(dfsapi[['kode_kabupaten_kota', 'tahun', 'daging_sapi']], on=['kode_kabupaten_kota', 'tahun'], how='left')
dfm = dfm.merge(dfkubis[['kode_kabupaten_kota', 'tahun', 'kubis']], on=['kode_kabupaten_kota', 'tahun'], how='left')
dfm = dfm.merge(dfsusu_sapi[['kode_kabupaten_kota', 'tahun', 'susu_sapi']], on=['kode_kabupaten_kota', 'tahun'], how='left')
dfm = dfm.merge(dftuna[['kode_kabupaten_kota', 'tahun', 'tuna']], on=['kode_kabupaten_kota', 'tahun'], how='left')
```

### PROSES

#### Barplot untuk melihat rata-rata produksi daging per kota/kabupaten

```python
dfp = dfm.groupby(['nama_kabupaten_kota'])[['daging_ayam', 'daging_babi', 'daging_domba', 'daging_kelinci', 'daging_sapi']].mean()
dfp.plot(kind='bar')
plt.title('Rata-rata Produksi Daging per Kota/Kabupaten')
plt.xlabel('Kota/Kabupaten')
plt.ylabel('Rata-rata Produksi')
plt.show()
```

Dapat kita lihat diatas kebanyakan kota/kabupaten di Jawa Barat rata-rata memproduksi daging ayam lebih banyak daripada daging lain dengan yang terbanyak adalah Kabupaten Bogor. Tetapi ada juga beberapa kota/kabupaten yang produksi sapinya lebih banyak daripada ayam seperti Kota Bandung dan Kota Bogor.

#### BARPLOT untuk melihat rata - rata dari produksi daging di jawa barat

```python
variables_to_plot = ['daging_ayam', 'daging_babi', 'daging_domba', 'daging_kelinci', 'daging_sapi']

# Hitung rata-rata konsumsi daging untuk setiap variabel
mean_values = dfm[variables_to_plot].mean()

# Buat bar plot
plt.figure(figsize=(12, 6))
plt.bar(mean_values.index, mean_values)
plt.title('Rata-rata Produksi Daging')
plt.xlabel('Variabel Daging')
plt.ylabel('Rata-rata Produksi')
plt.xticks(rotation=45)  # Untuk memutar label sumbu x

# Tampilkan plot
plt.show()
```

Dapat kita lihat dari visualisasi data di atas rata - rata produksi daging ayam jauh lebih tinggi dari pada produksi daging yang lain, disusul daging sapi, daging domba, daging kelinci dan daging babi.

#### Line Plot Pengeluaran Per Kapita Per tahun

```python
# Membuat line plot menggunakan Seaborn
plt.figure(figsize=(8, 6))
sns.lineplot(data=dfm, x='tahun', y='jumlah_pengeluaran_per_kapita', marker='o', color='b');
plt.title('Pengeluaran Per Kapita per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Pengeluaran Per Kapita')

plt.tight_layout()
plt.show()
```

Dapat Kita lihat di line plot diatas pengeluaran per kapita sempat naik di tahun 2019, tetapi dapat kita lihat dikarenakan di 2020 ada covid 19 pengeluaran per kapita menurun.

#### Line Plot Memvisualisasikan data produksi daging Per tahun

#### BARPLOT untuk melihat produksi Ikan Tuna Kabupaten/Kota

```python
dfm = dfm.sort_values(by='tuna', ascending=False)

# Membuat bar plot horizontal
plt.figure(figsize=(10, 6))  # Mengatur ukuran plot
plt.barh(dfm['nama_kabupaten_kota'], dfm['tuna'], color='skyblue')
plt.xlabel('Produksi Ikan Tuna')  # Label sumbu x
plt.ylabel('Nama Kabupaten/Kota')  # Label sumbu y
plt.title('Produksi Ikan Tuna per Kabupaten/Kota')  # Judul plot

# Menampilkan plot
plt.show()
```

Setelah memvisualisasikan data dengan menggunakan barplot dapat kita lihat produksi ikan tuna paling banyak di daerah - daerah yang mempunyai wilayah laut seperti Kabupaten Garut, Kabupaten Sukabumi, Kota Banjar dan Kabupaten Sumedang. Tetapi ada sesuatu sangat menarik yang di visualisasi data di atas, kota Bandung yang tidak mempunyai wilayah laut ternyata juga memproduksi ikan Tuna bahkan jumlahnya di atas Kabupaten Sumedang.
